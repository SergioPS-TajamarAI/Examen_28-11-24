# Encontrando insights de la UEFA Champions League

La Liga de Campeones de la UEFA, a menudo llamada Liga de Campeones, es una competencia anual de fútbol por excelencia que cautiva a los fanáticos de todo el mundo. Establecida en 1955 como la Copa de Clubes Campeones de Europa, evolucionó a la Liga de Campeones de la UEFA en 1992, ampliando su atractivo. El formato moderno cuenta con 32 equipos de clubes de primer nivel seleccionados en función de su desempeño en la liga nacional, lo que aumenta la intriga.  

Este evento electrizante trasciende los deportes y se convierte en una celebración de la unidad, la cultura y el orgullo nacional. Los fanáticos, vestidos con los colores de sus países, crean una atmósfera eléctrica, lo que hace que el torneo sea tanto sobre los espectadores como sobre los jugadores. Financieramente, la Liga de Campeones es un salvavidas para los clubes, ya que aumenta los ingresos y ofrece oportunidades transformadoras. Sin embargo, genera debates sobre las disparidades de riqueza en el fútbol europeo.  

La Liga de Campeones es sinónimo de rivalidades históricas, triunfos de los desvalidos y brillantez individual. Para los jugadores, representa la cima de su carrera, mientras que para los fanáticos, es un fenómeno cultural. El himno y los rituales icónicos enriquecen la experiencia futbolística. En 200 palabras, la UEFA Champions League es el epítome de la excelencia del fútbol europeo, que ofrece momentos inolvidables, recompensas económicas y un impacto cultural único, con 32 clubes de primer nivel que se suman a su atractivo.

1. Encuentra los 3 equipos que anotaron más goles jugando en su estadio en la UEFA Champions League 2020-21. El resultado debe contener dos columnas: TEAM_NAME_HOME y TEAM_HOME_SCORE ordenadas en orden descendente de TEAM_HOME_SCORE. Guarda la consulta (o el notebook) como TEAM_HOME_WITH_MOST_GOALS.

```
from pyspark.sql.functions import sum
df_home_goals = df_2021.select("TEAM_NAME_HOME", "TEAM_HOME_SCORE")
# Agrupa por equipo local y suma los goles
df_team_home_goals = df_home_goals.groupBy("TEAM_NAME_HOME").agg(sum("TEAM_HOME_SCORE").alias("TOTAL_HOME_GOALS"))

# Ordena los resultados y selecciona los 3 equipos con más goles
df_top_3_home_goals = df_team_home_goals.orderBy("TOTAL_HOME_GOALS", ascending=False).limit(3)

# Muestra el resultado
df_top_3_home_goals.show()
```

2. Encuentra el equipo con posesión mayoritaria la mayor cantidad de veces durante la UEFA Champions League 2021-22. El resultado debe incluir dos columnas: TEAM_NAME y GAME_COUNT, que es la cantidad de veces que el equipo tuvo posesión mayoritaria durante un partido de fútbol. Guarda esta consulta (o el notebook) como TEAM_WITH_MAJORITY_POSSESSION

```
from pyspark.sql import functions as F

# Supongamos que el dataframe se llama df

# Primero, convertimos las columnas de posesión a tipo numérico para poder compararlas
# Convertimos las columnas de posesión (que son strings) a enteros
df_2021 = df_2021.withColumn("POSSESSION_HOME_NUM", F.regexp_replace("POSSESSION_HOME", "%", "").cast("float"))
df_2021 = df_2021.withColumn("POSSESSION_AWAY_NUM", F.regexp_replace("POSSESSION_AWAY", "%", "").cast("float"))

# Luego, para cada fila, comparamos cuál equipo tiene más posesión
df_2021 = df_2021.withColumn(
    "TEAM_WITH_MAJOR_POSSESSION",
    F.when(F.col("POSSESSION_HOME_NUM") > F.col("POSSESSION_AWAY_NUM"), F.col("TEAM_NAME_HOME"))
     .when(F.col("POSSESSION_HOME_NUM") < F.col("POSSESSION_AWAY_NUM"), F.col("TEAM_NAME_AWAY"))
     .otherwise(F.lit(None))  # En caso de empate, asignamos None o podemos elegir un equipo por defecto
)

# Ahora agrupamos por el equipo con mayor posesión y contamos cuántos partidos ha ganado en posesión
result = df_2021.groupBy("TEAM_WITH_MAJOR_POSSESSION").agg(
    F.count("TEAM_WITH_MAJOR_POSSESSION").alias("GAME_COUNT")
)

# Encontramos el valor máximo de GAME_COUNT
max_game_count = result.agg(F.max("GAME_COUNT")).collect()[0][0]

# Filtramos para mostrar solo el equipo(s) con el mayor número de partidos con posesión mayoritaria
result_filtered = result.filter(F.col("GAME_COUNT") == max_game_count)

# Mostramos el resultado
result_filtered.show()
```

3. Encuentra la lista de equipos de cada fase del juego que ganaron el duelo en un partido pero terminaron perdiendo el juego en el Campeonato de la UEFA 2022-23. El resultado debe contener dos columnas: STAGE y TEAM_LOST. Guarda la consulta (o el notebook) como TEAM_WON_DUEL_LOST_GAME_STAGE_WISE.

```
from pyspark.sql import functions as F

# Supongamos que el dataframe se llama df_2022

# 1. Determinar quién ganó el partido utilizando los goles
df_2022 = df_2022.withColumn(
    "who_won",
    F.when(F.col("TEAM_HOME_SCORE") > F.col("TEAM_AWAY_SCORE"), "HOME")
     .when(F.col("TEAM_HOME_SCORE") < F.col("TEAM_AWAY_SCORE"), "AWAY")
     .otherwise("DRAW")
)

# 2. Filtrar los equipos que ganaron uno de los partidos (HOME o AWAY) pero no avanzaron
# Suposición: Si el equipo no pasó de ronda, es el equipo que perdió en la segunda parte de la eliminatoria
df_2022 = df_2022.withColumn(
    "TEAM_WON_ONE_BUT_LOST_ELIMINATION",
    F.when(
        ((F.col("who_won") == "HOME") & (F.col("TEAM_HOME_SCORE") > F.col("TEAM_AWAY_SCORE"))) |
        ((F.col("who_won") == "AWAY") & (F.col("TEAM_AWAY_SCORE") > F.col("TEAM_HOME_SCORE"))),
        F.coalesce(F.col("TEAM_NAME_HOME"), F.col("TEAM_NAME_AWAY"))
    ).otherwise(F.lit(None))
)

# Filtrar para solo seleccionar los equipos que ganaron un partido pero no avanzaron
result = df_2022.filter(F.col("TEAM_WON_ONE_BUT_LOST_ELIMINATION").isNotNull())

result.display()

# 3. Agrupar por fase (STAGE) y mostrar el equipo y fase
result_stage_wise = result.groupBy("STAGE", "TEAM_WON_ONE_BUT_LOST_ELIMINATION").agg(
    F.count("TEAM_WON_ONE_BUT_LOST_ELIMINATION").alias("GAME_COUNT")
)

# 4. Mostrar el resultado
result_stage_wise.show()
```
